```python
def convert_examples_to_features(example_batch):
    input_encodings = tokenizer(example_batch['dialogue'] , max_length = 1024, truncation = True )

    with tokenizer.as_target_tokenizer():
        target_encodings = tokenizer(example_batch['summary'], max_length = 128, truncation = True )

    return {
        'input_ids' : input_encodings['input_ids'],
        'attention_mask': input_encodings['attention_mask'],
        'labels': target_encodings['input_ids']
    }
```

```python
dataset_samsum_pt = dataset_samsum.map(convert_examples_to_features, batched = True)
```

```python
dataset_samsum_pt["train"]
```

## **Training**

```python
from transformers import DataCollatorForSeq2Seq, TrainingArguments, Trainer
import transformers
```

```python
seq2seq_data_collator = DataCollatorForSeq2Seq(tokenizer, model=model_pegasus)
```

```python
trainer_args = TrainingArguments(
    output_dir='pegasus-samsum',
    num_train_epochs=4,
    per_device_train_batch_size=1,
    per_device_eval_batch_size=1,
    weight_decay=0.01,
    logging_steps=10,
    save_steps=1000,
    gradient_accumulation_steps=16
)
```

```python
trainer = Trainer(
    model=model_pegasus,
    args=trainer_args,
    tokenizer=tokenizer,
    data_collator=seq2seq_data_collator,
    train_dataset=dataset_samsum_pt["train"],
    eval_dataset=dataset_samsum_pt["validation"]
)
```

```python
trainer.train()
```

```python
trainer.save_model("pegasus-samsum-model")
tokenizer.save_pretrained("pegasus-samsum-tokenizer")
```

```python
def generate_batch_sized_chunks(list_of_elements, batch_size):
    """Split the dataset into smaller batches that we can process simultaneously
    Yield successive batch-sized chunks from list_of_elements."""
    for i in range(0, len(list_of_elements), batch_size):
        yield list_of_elements[i : i + batch_size]
```

```python
def calculate_metric_on_test_ds(dataset, metric, model, tokenizer,
                              batch_size=16, device=device,
                              column_text="dialogue",
                              column_summary="summary"):
    article_batches = list(generate_batch_sized_chunks(dataset[column_text], batch_size))
    target_batches = list(generate_batch_sized_chunks(dataset[column_summary], batch_size))

    for article_batch, target_batch in tqdm(
        zip(article_batches, target_batches), total=len(article_batches)):

        inputs = tokenizer(article_batch, max_length=1024, truncation=True,
                        padding="max_length", return_tensors="pt")

        summaries = model.generate(input_ids=inputs["input_ids"].to(device),
                        attention_mask=inputs["attention_mask"].to(device),
                        length_penalty=0.8, num_beams=8, max_length=128)

        decoded_summaries = [tokenizer.decode(s, skip_special_tokens=True,
                                clean_up_tokenization_spaces=True)
               for s in summaries]

        decoded_summaries = [d.replace("", " ") for d in decoded_summaries]

        metric.add_batch(predictions=decoded_summaries, references=target_batch)

    score = metric.compute()
    return score
```

```python
from datasets import load_metric
rouge_metric = load_metric('rouge')
rouge_names = ["rouge1", "rouge2", "rougeL", "rougeLsum"]
```

```python
test_samples = dataset_samsum['test'][:20]  # Using first 20 samples for faster evaluation
score = calculate_metric_on_test_ds(
    test_samples, rouge_metric, trainer.model, tokenizer,
    batch_size=2, column_text='dialogue', column_summary='summary'
)
```

```python
rouge_dict = dict((rn, score[rn].mid.fmeasure) for rn in rouge_names)
import pandas as pd
results_df = pd.DataFrame(rouge_dict, index=['pegasus-samsum'])
print(results_df)
```
